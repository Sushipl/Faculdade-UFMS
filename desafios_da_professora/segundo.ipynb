```python

n = 0

while True:

    try:

        n = int(input("Dígite um número maior que zero: "))

        if n > 0:

            break

    except:

        print("Dígite um número inteiro!")

ini = 0

soma = 0

while soma < n:

    soma += ini

    ini += 1


if soma == n:

    print('É triangular')

else:

    print("Não é triangular")

```

```python
n = 0

while True:

    try:

        n = int(input("Dígite um número maior que zero: "))

        if n > 0:

            break

    except:

        print("Dígite um número inteiro!")

        
soma=0

while n > 0:

    soma += n

    n -= 1

print(soma)
```

```python
while True:

    try:
        cpf = input("Dígite os 9 primeiros dígitos do seu CPF(apenas os números): ")

        if len(cpf) == 9 and int(cpf):

            break

        print("Dígite corretamente")

    except:

        print("Apenas os números!")

listdicpf = []

valcalc = []

for n in range(10,1,-1):

    valcalc.append(n)

for n in cpf:

    listdicpf.append(int(n))


calc = []

for pos in range(9):

    calc.append(listdicpf[pos]*valcalc[pos])

soma = 0

for n in calc:

    soma += n

soma, rest = soma//11, soma%11

if rest < 2:

    listdicpf.append(0)

else:

    listdicpf.append(11-rest)

valcalc = [11]+ valcalc


calc = []

for pos in range(10):

    calc.append(listdicpf[pos]*valcalc[pos])

soma = 0

for n in calc:

    soma += n

soma, rest = soma//11, soma%11

if rest < 2:

    listdicpf.append(0)

else:

    listdicpf.append(11-rest)

cpf = ""

for n in listdicpf:

    cpf += str(n)

print(cpf)
```
